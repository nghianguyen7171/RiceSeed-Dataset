```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
Data = pd.read_csv('riceseeds.csv')
Data.head()
Data.shape
```

```python
#Choose 6 motil species
df = pd.DataFrame(Data[Data['Species'].isin(['BC15','BQ10','NH92','NT16','PC10','VietHuong8'])])
df.reset_index(drop=True, inplace=True)
df
```

```python
X = df.values[:, 8:263]
Y = df.values[:, 1]
X.shape
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.2, random_state = 96)
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
X_train.shape
```

```python
#PCA
#from sklearn.decomposition import PCA # Principal Component Analysis module
#pca = PCA(n_components=2)
#X_PCA = pca.fit_transform(X)
#pca.fit(X_train)
#y_pred = lda.predict(X_test)
#print("Accuracy is"), accuracy_score(y_test,y_pred)*100
```

```python
#RF
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 200, criterion = 'entropy', max_depth = 5, min_samples_leaf = 1, random_state = 40)
classifier.fit(X_train,y_train)
```

```python
Y_pred = classifier.predict(X_test)
print("Accuracy is"), accuracy_score(y_test,Y_pred)*100
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components=2)
#Xlda = lda.fit_transform(X,Y)
```

```python
lda.fit(X_train,y_train)
y_pred = lda.predict(X_test)
print("Accuracy is"), accuracy_score(y_test,y_pred)*100
```

```python
#PCA
from sklearn.decomposition import PCA
pca = PCA(n_components= 2)
principalComponents = pca.fit_transform(X_train)
PCA_DF = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
PCA_DF.shape
```
